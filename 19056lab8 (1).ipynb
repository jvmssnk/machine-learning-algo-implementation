```python
import pandas as pd
import math
 
data = {'age': ['<=30','<=30','31..40','>40','>40','>40','31..40','<=30','<=30','>40','<=30','31..40','31..40','>40',],
        'income': ['high','high','high','medium','low','low','low','medium','low','medium','medium','medium','high','medium'],
        'students': ['no','no','no','no','yes','yes','yes','no','yes','yes','yes','no','yes','no'],
        'credits_rating':['fair','excellent','fair','fair','fair','excellent','excellent','fair','fair','fair','excellent','excellent','fair','excellent'],
        'buys_computer':['no','no','yes','yes','yes','no','yes','no','yes','yes','yes','yes','yes','no']
        }
df = pd.DataFrame(data, columns= ['age', 'income','students','credits_rating','buys_computer'])

print (df)
```

```python
k=df['buys_computer']
c=0
kp=0
for i in k:
    if i=='yes':
        c=c+1
    elif i=='no':
        kp=kp+1
entro=-(c/len(k))*math.log(c/len(k),2)
entro1=-(kp/len(k))*math.log(kp/len(k),2)
totq=entro+entro1
totq
finall=[]
```

```python
l=[]
lis=[]
res=[]
def entrop(tot,er):
    l,k1,k2,tsi,si=er[0],er[1],er[2],er[3],er[4]
    kfr=-(si/tsi)
    if k1==0:
        kfr1=0
    else:
        kfr1=-(k1/si)*math.log((k1/si),2)
    if k2==0:
        kfr2=0
    else:
        kfr2=-(k2/si)*math.log((k2/si),2)
    return -1*kfr*(kfr1+kfr2)
    
def mycommon(se,v):
    kg=list(df[df[se]==v][se])
    kio=list(df[df[se]==v]['buys_computer'])
    yes=kio.count('yes')
    no=kio.count('no')
    p=len(k)
    p1=len(kg)
    return (kg,yes,no,p,p1)
```

```python
l=[]
se='age'
lis=[]
res=[]
op=(df[se].unique())
for i in op:
    lis.append(i)
for i in lis:
    l.append(mycommon(se,i))
print(l)
for j in l:
    res.append(entrop(totq,j))
finall.append((totq-sum(res)))
```

```python
import os
import cv2
```

```python
images_path='C:\\Users\\josyu\\Downloads\\img_align_celeba\\img_align_celeba'
```

```python
def processimage(path):
    print(path)
    img = cv2.imread(path)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    newimg = cv2.resize(gray, (0, 0), fx = 0.1, fy = 0.1)
    redimg=newimg.reshape(396,)
    return redimg
```

```python
classa=[]
for i in range(0,100):
    temp=processimage(images_path+"/"+os.listdir(images_path)[i])
    classa.append(temp)
```

```python
classa=np.array(classa)
DATAa=pd.DataFrame(classa)
```

```python
DATAa["Label"]=[0]*100
```

```python
DATAa.head()
```

```python
classb=[]
for i in range(100,200):
    temp=processimage(images_path+"/"+os.listdir(images_path)[i])
    classb.append(temp)
```

```python
classb=np.array(classb)
DATAb=pd.DataFrame(classb)
```

```python
DATAb["Label"]=[1]*100
```

```python
DATAb.head()
```

```python
final_data=pd.concat([DATAa,DATAb])
```

```python
final_data.shape
```

```python
x=final_data.drop("Label",axis=1)
y=final_data["Label"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```
