```python
def processimage(path):
    print(path)
    img = cv2.imread(path)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    newimg = cv2.resize(gray, (0, 0), fx = 0.1, fy = 0.1)
    redimg=newimg.reshape(396,)
    return redimg
```

```python
def PCA_implementation(data , components):
    mean_data = data - np.mean(data , axis = 0)
    covariance_matrix = np.cov(mean_data , rowvar = False)
    eigen_values , eigen_vectors = np.linalg.eigh(covariance_matrix)
    sorted_data = np.argsort(eigen_values)[::-1]
    sorted_eigenvalue = eigen_values[sorted_data]
    sorted_eigenvectors = eigen_vectors[:,sorted_data]
    selected_eigen_vectors = sorted_eigenvectors[:,0:components]
    final_data = np.dot(selected_eigen_vectors.transpose() , mean_data.transpose()).transpose()
    #Reconstructing 
    x = np.matmul(final_data,selected_eigen_vectors.T) + np.mean(data,axis=0)
    return final_data,selected_eigen_vectors,x
```

```python
def forward_feature_selection(x,y,op,acc,memo):
    while len(op)<=396:
        print(f"length of op is {len(op)}")
        top=op
        temp=[i for i in range(396) if i not in op]
        select=None
        for i in temp:
            top.append(i)
            model=LogisticRegression()
            model.fit(x[top],y)
            sc=model.score(x[top],y)
            if acc<sc:
                acc=sc
                select=i
            top.pop()
        if select==None:
            return memo
        else:
            op.append(select)
        model=LogisticRegression()
        model.fit(x[op],y)
        sc=model.score(x[op],y)
        memo[str(op)]=sc
    return memo
```
