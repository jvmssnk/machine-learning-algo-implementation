```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
sns.set()
```

```python
import os
import cv2
```

```python
images_path='C:\\Users\\josyu\\Downloads\\img_align_celeba\\img_align_celeba'
```

```python
def processimage(path):
    print(path)
    img = cv2.imread(path)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    newimg = cv2.resize(gray, (0, 0), fx = 0.1, fy = 0.1)
    redimg=newimg.reshape(396,)
    return redimg
```

```python
classa=[]
for i in range(0,100):
    temp=processimage(images_path+"/"+os.listdir(images_path)[i])
    classa.append(temp)
```

```python
classa=np.array(classa)
DATAa=pd.DataFrame(classa)
```

```python
DATAa.head()
```

```python
classb=[]
for i in range(100,200):
    temp=processimage(images_path+"/"+os.listdir(images_path)[i])
    classb.append(temp)
```

```python
classb=np.array(classb)
DATAb=pd.DataFrame(classb)
```

```python
DATAb.head()
```

```python
final_data=pd.concat([DATAa,DATAb])
```

```python
final_data.shape
```

```python
x=final_data[[0]]
y=final_data[1]
```

```python
y.shape
```

```python
x.shape
```

```python
plt.scatter(x,y)
plt.xlabel('images', fontsize = 20)
plt.ylabel('labels', fontsize = 20)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
regr = linear_model.LinearRegression()
regr.fit(X_train,y_train)
y_pred = regr.predict(X_test)
print('Coefficients: \n', regr.coef_)
print('Mean squared error: %.2f'% mean_squared_error(y_test,y_pred))
print('Coefficient of determination: %.2f'% r2_score(y_test,y_pred))
plt.scatter(X_test,y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xticks(())
plt.yticks(())
plt.show()
```
